```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
salary=pd.read_csv('Salary_Data.csv')
salary
```

```python
salary.isnull()
```

```python
x=salary[['YearsExperience']]
x
```

```python
y=salary[['Salary']]
y
```

```python
plt.scatter(x,y,color='black',marker='*')
plt
```

```python
#Linear regression and fit the model
classifier = LinearRegression()
classifier.fit(x,y)
```

```python
#Predict for X dataset
y_pred = classifier.predict(x)
y_pred
```

```python
classifier.coef_
```

```python
classifier.intercept_
```

```python
9449.96232146*10.5+25792.20019867
```

```python
salary.head(3)
```

```python
salary['prediction']=y_pred
```

```python
salary
```

```python
classifier.score(x,y)
```

```python
plt.xlabel('Years of Experience', fontsize=20)
plt.ylabel('salary' ,fontsize=20)
plt.scatter(salary.YearsExperience,salary.Salary,color='black',marker='*')
plt.plot(salary.YearsExperience,y_pred,color='blue')
```

```python
from sklearn.metrics import r2_score
sol=r2_score(y,y_pred)
```
