```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
win = pd.read_csv("wine.csv")
win.head()
```

```python
wine_data = win.iloc[:,1:]
wine_data
# Converting into numpy array
WINE = wine_data.values
WINE
```

```python
# Normalizing the numerical data 
win_normal = scale(WINE)
win_normal
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(win_normal)
pca_values 
```

```python
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
 pca_values[:,0:1]
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3']), win[['Type']]], axis = 1)
finalDf
```

# BY USING KMEANS


```python
from sklearn.cluster import KMeans
```

```python
data2=finalDf
data2
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(data2)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=52)
k_hm=clusters_new.fit_predict(data2)
```

```python
k_hm
```

```python
kmeans_cluster=pd.Series(k_hm)
kmeans_cluster
wine1['kclust']=kmeans_cluster # creating a  new column and assigning it to new column 
wine1
```

```python
wine1.iloc[:,1:14].groupby(wine1.kclust).mean()
```

```python
wine1.plot(x="Alcohol",y ="Ash",c=k_hm,kind="scatter",s=50 ,cmap=plt.cm.cividis_r) 
plt.title('Clusters using KMeans')
```
