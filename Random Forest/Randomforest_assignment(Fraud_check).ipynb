```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_csv('Fraud_Check.csv')
data.head()
```

```python
data['Tax_New']=pd.cut(data['Taxable.Income'],bins=[0,30000,99619+1],labels=['Risky','Good'])
df=data.drop(['Taxable.Income'],axis=1)
df
```

```python
df['Tax_New'].value_counts()
```

```python
data1 = pd.get_dummies(df)
data1
```

```python
data1.drop(['City.Population'],axis=1,inplace=True)
data1.head(2)
```

```python
data2=data1.drop(['Tax_New_Risky'],axis=1)
data2
```

```python
X=data2.drop(['Tax_New_Good'],axis=1)
Y=data2['Tax_New_Good']
```

```python
num_trees = 100
max_features = 3
kfold = KFold(n_splits=10, random_state=7)
```

```python
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean()) 
```
